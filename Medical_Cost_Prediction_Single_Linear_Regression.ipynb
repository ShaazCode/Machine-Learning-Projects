Splitting target variable

```python
X = df.drop('annual_medical_cost', axis=1)
Y = df['annual_medical_cost']

# here we had removed X column from the dataset
```

Split the dataset into train , test and split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
# here the data had been splited into trainign and testing part
# 20% data is for testing whreas 80% for training
```

Model selection

```python
# import the model
```

```python
from sklearn.linear_model import LinearRegression
```

Create the model object

```python
model = LinearRegression()
```

Train the model

```python
model.fit(X_train, Y_train)
```

Model Prediction

```python
Y_pred = model.predict(X_test)
```

```python
print(Y_pred)
```

Compare with actual values

```python
print("Actual values:")
print(Y_test.values)

print("\nPredicted values:")
print(Y_pred)
```

```python
import pandas as pd

comparison = pd.DataFrame({
    'Age': X_test.values.flatten(),
    'Actual_Cost': Y_test.values,
    'Predicted_Cost': Y_pred
})

comparison
```
