# if-else

we're making use of else when the if condition is not getting true.
when none of the condition is getting true at that time we're making use of else part

```python
# when we want to check 2 condiiton at that time we're making use of else part
marks = 85

if marks <= 75:
  print("First Class")

else:
  print("Second class")
```

# If-elif-else


if we want to compare more than 2 condition then we're making use of if elif else part

```python
marks = 65

if marks >= 90:
  print("First class")

elif marks > 80:
  print("Second class")

elif marks >=70:
    print("Third class")

elif marks >= 60:
  print("Fourth class")

else:
  print("D class")
```

# **List**

```python
# List is used to store the data , and the data can be of any types
# O C D (properties)
# O -> ordered
# C -> changable
# D -> Duplicates

# in built dt RE EXTENSION OF varaible in var we can store single value bt  in inbuilt we can store multiple values
```

```python
# if we want to access any element then we can access it with the help of index location
```

```python
# List
# List follows O C D
# ordered: in same way we're getting the element
# changable: we can add, delete and updates the values

a = [10,20,30,40,50]
print("original list", a)

# add 60 in it at last position.  eg a = [10,20,30,40,50,60]
# append -> add at end
a.append(60)
print("Adding element at last ", a)

# add 70 at 3rd position.  eg a = [10,20,30,40,50,60]
# for accessing we can make use of index position
a.insert(3, 70)
# insert takes to paramete3r 1. where to add and what to add
print("Adding element at particualr position",a)

# Deletion of element
a.pop(4)# pop is used to delete the element by index
print("After popping", a)


# updating element
a[2] = 300
print("updating the list elemnt",a)
```

# **Tuples**

```python
# Tuples are used to store the data and data can be any type

# Properties : O Cx D
# tuples has only 2 properties ordered and duplicates , changing element is not allowed in tuples -> no add ,delete, update the value

a = (10,20,30,40,50)
print("Tuple values are ", a) # ordered


a = (10,20,30,40,50,50)  # duplicates
print(a)


# inside tuple we cannot add , delete athe element
```

```python
print(a[2])

# if we want to access any element then we can access it using index values
# if u want to access any value then we can wrap the index value in []
```

# **Set**

```python
# set is also used to store the data and the data can be in any format.

# Properties: Ox Dx
# set does not follow ordere property

a = {10,20,30,40}
print("set element is not in ordered", a)
```

```python

a = {10,20,30,10,40,40}
print("set element does not allow duplicates", a)

print(a[2])
```

# **While loop**

it will run till the condition is getting satisfy

It will keep doing until the condition is getting false

```python
# synatx:
# while condition
```

```python
# Task is print a number from 1 to 5

i = 1

while i<=5:
  print(i)
  i = i + 1
```

```python
# print a number from 10 to 1

num = 10

while num >=1:
  print(num)
  num = num - 1
```
